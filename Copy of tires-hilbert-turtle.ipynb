```python
def hilbert(turtle, iteration, size, angle):#Defining A
    if (iteration == 0):#If the iteration is 0, no line is drawn so we return
        return
    else:
        turtle.left(angle) #+ (turning left)
        hilbert(turtle, iteration - 1, size, -angle) #B (negative angle in A gives B)
        turtle.forward(size) #F
        turtle.left(-angle) #-
        hilbert(turtle, iteration - 1, size, angle) #A
        turtle.forward(size) #F
        hilbert(turtle, iteration - 1, size, angle) #A. We are using iteration - 1 to decrease the depth on each iteration until it reaches 0
        turtle.left(-angle) #-
        turtle.forward(size) #F
        hilbert(turtle, iteration - 1, size, -angle) #B
        turtle.left(angle)
```

```python
iteration = 2
size = 400/2 ** iteration
angle = 90
```

```python
turtle.initializeTurtle()
turtle.clearscreen()
turtle.setup(width=1000, height=500)
```

```python
def R(angle):
  theta = np.pi*angle/180
  return np.matrix([
      [np.cos(theta), -np.sin(theta)],
      [np.sin(theta), np.cos(theta)]
  ])
```
